```python
import cv2
import time 
video = cv2.VideoCapture("./stop_py.avi")


output_video = cv2.VideoWriter('output_video_py.avi', cv2.VideoWriter_fourcc(*'XVID'), 30, (400, 600))


timeframe = 0
new_timeframe = 0


while(video.isOpened()):
    
    success, frame = video.read()
    if not success : 
        break
    
    frame = cv2.resize(frame,(400,600))
    new_timeframe = time.time()
    fps = 1/(new_timeframe- timeframe)
    timeframe = new_timeframe
    
    fps = int(fps)
    
    cv2.putText(frame, str(fps), (8,80), cv2.FONT_HERSHEY_SIMPLEX, 3, (100,255,0), 4)
    cv2.imshow("vid", frame)
    
    output_video.write(frame)

    if cv2.waitKey(1) & 0xFF == ord('q'): 
        
        break

video.release()
output_video.release()
cv2.destroyAllWindows()

```
